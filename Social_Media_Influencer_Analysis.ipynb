```python
import pandas as pd
```

```python
data = pd.read_csv('C:/social_media/social media influencers-instagram june 2022 - june 2022.csv')
```

```python
data.head()
```

```python
data = data.rename({'Views avg.': 'Country'}, axis=1)
```

```python
data.shape
```

Clean data, convert strings to numbers

```python
data = data.drop_duplicates()
data.shape
```

```python
data.isna().sum()
```

```python
data = data.drop(columns=['Category_2'])
```

```python
data.columns
```

```python
data = data.dropna()
```

```python
data.head()
```

```python
data['Subscribers_count'] = data['Subscribers count'].replace({"K":"*1e3", "M":"*1e6"}, regex=True).map(pd.eval).astype(int)
data['Likes_avg'] = data['Likes avg'].replace({"K":"*1e3", "M":"*1e6"}, regex=True).map(pd.eval).astype(int)
data['Comments_avg.'] = data['Comments avg.'].replace({"K":"*1e3", "M":"*1e6"}, regex=True).map(pd.eval).astype(int)
```

```python
data.head()
```

```python
data = data.drop(columns=['Subscribers count','Likes avg','Comments avg.'])
```

```python
data.head()
```

Get average subscriber/follower count by category

```python
data.groupby('Category_1')['Subscribers_count'].mean()
```

Top influencers by category

```python
df = data.groupby('Category_1')['Subscribers_count'].idxmax().reset_index()
df.head()
```

```python
data.loc[df['Subscribers_count']].sort_values(by = 'Subscribers_count', ascending=False)
```

Define engagement metrics

```python
data['Engagement_rate']=((data['Likes_avg']+data['Comments_avg.'])/data['Subscribers_count'])*100
data.head()
```

Favored categories by country

```python
import numpy as np
table = pd.pivot_table(data, values='Engagement_rate', index=['Category_1'],
                     columns=['Country'], aggfunc=np.sum)
```

```python
table.idxmax()
```
